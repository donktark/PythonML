암게의 신체적 특성에 따른 부수체(숫게; sat)의 개수에 대한 데이터이다. </p>
실습을 위해 일단 width 변수만을 사용해서 부수체의 개수를 예측한다.

```python
wt_df = df[['sat','width']]
wt_df
```

```python
fit = sm.formula.glm('sat ~ width', family=sm.families.Binomial(), data=wt_df)
```
