```python
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import sklearn
import mglearn
from sklearn.svm import SVC, LinearSVC
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn import metrics
from sklearn.metrics import confusion_matrix, roc_auc_score, accuracy_score, f1_score, precision_recall_curve
from sklearn.metrics import roc_curve, average_precision_score, matthews_corrcoef
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
```

```python
path = "C://Users//32197453//Desktop//교과//4학년 2학기//기계학습이론//"
df = pd.read_csv(path + "/과제/2/pulsar_stars.csv")
# df.set_index("Drug", inplace=True)
df
```

```python
print("--------데이터셋 차원 확인")
print(df.shape)
print("--------결측값 있는 변수 확인")
missval = df.isnull().sum()[df.isnull().sum()!=0]
print(missval)
missval_name = missval.index #결측값 있는 변수 이름 벡터로 추출
```

```python
#X,Y 데이터셋 분리
X = df.drop("target_class", axis=1)
y = df["target_class"]
print("--------반응변수 값 확인")
print(y.unique())
```

```python
#학습용 검증용 데이터셋 분리
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=7453)
```

```python
#파이프라인 작업 저장
poly_svm = Pipeline([
    ("scaler", StandardScaler()),
    ("svm_clf", SVC(C=100, degree=3, gamma=0.1, kernel="poly"))
])
```

변수가 고차원이고 반응변수가 0,1로 이진분류 되어있으므로 Nonlinear SVM에서 kernel을 polynomial을 사용하여 분류한다.

```python
#모델 적합
poly_svm.fit(x_train,y_train)
y_pred = poly_svm.predict(x_test)
```

```python
#실제 데이터와 성능 비교

print(confusion_matrix(y_test, y_pred)); print()

#Accuracy
print("Accuracy :", accuracy_score(y_test, y_pred))

#AUC
print("AUC :", roc_auc_score(y_test, y_pred))

#AUPRC
print("AUPRC :", average_precision_score(y_test, y_pred))

#F1-score
print("F1-Score :", f1_score(y_test, y_pred))

#MCC
print( "MCC :", matthews_corrcoef(y_test, y_pred))
```

학습 결과 C=100, $\gamma$ = 0.1, 차수=3 일때 좋은 성능을 보인다.
